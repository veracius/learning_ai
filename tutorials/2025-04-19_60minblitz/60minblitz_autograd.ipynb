```python
import torch
from torchvision.models import resnet18, ResNet18_Weights

model = resnet18(weights=ResNet18_Weights.DEFAULT)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
print(data)
```

```python
prediction = model(data) #send random image through model
```

```python
#compute how much model output deviates from the randomly generated labels
loss = (prediction - labels).sum()
loss.backward()
print(loss)
```

```python
#load stochastic gradient descent algo to adjust model parameters based on loss fn/backprop
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
#step initiates gradient descent step
optim.step()
```

```python
#autograd:
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)

Q = 3*a**3 - b**2
```

```python
#tensor has a .backward() function that computes gradient (partial deriv of each input, and stores result of current value in .grad tensor attribute
external_grad = torch.tensor([1., 1.])
Q.backward(gradient=external_grad)

print(9*a**2 == a.grad)
print(-2*b == b.grad)
print(external_grad)
```

```python
from torch import nn, optim
model = resnet18(weights=ResNet18_Weights.DEFAULT)

for param in model.parameters():
    param.requires_grad = False #freeze all the parameters

#resent: classifier is last linear layer model.fc; add a new unfrozen linear layer to act as classifier
model.fc = nn.Linear(512, 10)

#optimize only the classifier:
optimizer = optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
optimizer.step()
```
