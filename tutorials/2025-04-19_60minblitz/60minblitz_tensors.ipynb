```python
import torch
import numpy as np
print(torch.cuda.is_available())
```

```python
data = [[1,2], [1,2]]
x_data = torch.tensor(data)
print(data)
print(x_data)
```

```python
#bridge NumPy array to tensor
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
backtonp = x_np.numpy() #I assume .detach().cpu().numpy is if the tensor is in some kind of chain?

print(np_array)
print(x_np)
print(backtonp)
```

```python
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Random Tensor: \n {x_rand} \n")

x_rand = torch.rand_like(x_data, dtype=torch.double)
print(f"Random Tensor: \n {x_rand} \n")
```

```python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
altones_tensor = torch.ones_like(ones_tensor)
zeros_tensor = torch.zeros(shape)

print(f"(Random Tensor:\n{rand_tensor}\n")
print(f"(Ones Tensor:\n{ones_tensor}\n")
print(f"(Ones (alt) Tensor:\n{altones_tensor}\n")
print(f"(Zeros Tensor:\n{zeros_tensor}")
```

```python
tensor = torch.rand(3,4)
print(f"Tensor shape: {tensor.shape}")
print(f"Tensor data type: {tensor.dtype}")
print(f"Tensor stored on device: {tensor.device}")
```

```python
#tensor = torch.rand(4,4)
tensor = torch.ones(4,4)
tensor[:, 1] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
t2 = torch.cat([tensor, tensor, tensor], dim=0) #why does this have to be in range [-2, 1]?
print(t2)
```

```python
#elementwise product
print(f"tensor.mul(tensor) \n{tensor.mul(tensor)} \n") #function call off tensor object
print(f"tensor*tensor\n{tensor * tensor}") #operator on tensor objects
```

```python
#matrix mult
tensor.matmul(tensor.T)
tensor @ tensor.T
```

```python
print(tensor, "\n")
tensor.add_(5) #"in-place" operations are per-element of tensor. Can cause issues with derivatives
print(tensor)
```

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(n)
```

```python
#changes to numpy reflect in tensor?
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```

```python
#do changes in tensor reflect in numpy? --> yes
t.sub_(1)
print(f"t: {t}")
print(f"n: {n}")
```
